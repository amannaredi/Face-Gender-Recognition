```python
#lets crop all images

def crop_faces(path , gender , i):
    img = cv2.imread(path)
    face = Haar.detectMultiScale(img , 1.5 , 5)
    for x,y,h,w in face :
        roi = img[y:y+h , x:x+w]
        
        if gender == "male" :
            cv2.imwrite("./data/male_crop/{}_{}.png".format(gender , i) , roi)
        else :
            cv2.imwrite("./data/female_crop/{}_{}.png".format(gender , i) , roi)    
        
```

```python
for i , path in enumerate(female_path):
    try :
        crop_faces(path , "female" , i)
        print("{}/{} processed successfully".format(i , len(female_path)))
      
    except :
        print("{}/{} not processed successfully".format(i , len(female_path)))
```

```python
for i , path in enumerate(male_path):
    try :
        crop_faces(path , "male" , i)
        print("{}/{} processed successfully".format(i , len(male_path)))
      
    except :
        print("{}/{} not processed successfully".format(i , len(male_path)))
```

### 60% & 40% , That is almost Balanced data

```python
# Resize data into 224*244 
img_re_x = []
label = []
def resize_images(path , gender):
    img = cv2.imread(path)
    if img.shape[0] < 224 :
        img_re = cv2.resize(img, (224,224) , cv2.INTER_CUBIC) #enlarge
    else :
        img_re = cv2.resize(img , (224,224) , cv2.INTER_AREA) #shrink
        
    img_re_x.append(img_re)
    label.append(gender)
    print("Process successfully completed")
```

```python
gender = data["gender"]
```

```python
gender = list(gender)
```

```python
path = data["Path"]
```

```python
path = list(path)
```

```python
for path , gender in zip(path , gender):
    resize_images(path , gender)
```
